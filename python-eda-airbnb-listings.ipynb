**4. Data Cleaning**

```python
data.isnull().sum()

data.dropna(inplace=True)
data.isnull().sum()
```

```python
# Dealing with duplicate rows
data.duplicated().sum()

# Deleting all duplicated rows

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
# Type casting
# changing Data types
data.dtypes

data['id'] = data['id'].astype(object)
data.dtypes

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

**EDA**

**5. Data Analysis**

**Univariate Analysis**

```python
# Calcutlating IQR
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]

# Now plot the cleaned prices
sns.boxplot(data=df, x='price')
```

```python
#Price Distribution
plt.Figure(figsize=(8,5))
sns.histplot(data=df ,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
plt.Figure(figsize=(6,3))
sns.histplot(data=df ,x='availability_365')
plt.title('availability_365')
plt.ylabel("Frequency")
plt.show()
```

```python
# Average price per bed

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Feature Engineering**

```python
#price per bed  calculation

df['price per bed']= df['price']/df['beds']
df.head()
```
